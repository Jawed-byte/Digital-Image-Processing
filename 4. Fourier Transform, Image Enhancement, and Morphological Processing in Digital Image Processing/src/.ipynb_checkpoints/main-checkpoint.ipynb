**1.1: DFT 1**

```python
import numpy as np
import librosa
import matplotlib.pyplot as plt

def custom_dft(signal):
    signal_length = signal.shape[-1]
    result = np.zeros(signal_length, dtype=np.complex128)
    indices = np.arange(signal_length)

    for freq_index in range(signal_length):
        result[freq_index] = np.sum(signal * np.exp(-2j * np.pi * freq_index * indices / signal_length))

    return result

audio_file = '/content/1.wav'
audio_signal, sample_rate = librosa.load(audio_file, sr=None)

dft_result = custom_dft(audio_signal)

signal_length = len(dft_result)
frequency_bins = np.arange(signal_length) * (sample_rate / signal_length)

positive_length = signal_length // 2
magnitude_values = np.abs(dft_result[:positive_length])
frequencies = frequency_bins[:positive_length]

plt.figure(figsize=(12, 6))
plt.plot(frequencies, magnitude_values)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Spectrum')
plt.grid(True)
plt.show()
```

**1.2: DFT 2**

```python
import numpy as np

def dft2(input_signal):

    original_dims = np.array(input_signal.shape)

    padding_required = np.power(2, np.ceil(np.log2(original_dims)).astype(int)) - original_dims
    padded_signal = np.pad(input_signal, ((0, padding_required[0]), (0, padding_required[1])), 'constant')

    transformed_signal = np.zeros(padded_signal.shape, dtype=np.complex128)

    for row in range(padded_signal.shape[0]):
        transformed_signal[row, :] = custom_dft(padded_signal[row, :])

    for col in range(padded_signal.shape[1]):
        transformed_signal[:, col] = custom_dft(transformed_signal[:, col])

    return transformed_signal
```

**1.3: FFT 1**

```python
import numpy as np

def custom_fft(input_signal):

    signal_length = len(input_signal)

    required_size = 2 ** np.ceil(np.log2(signal_length)).astype(int)
    padded_signal = np.pad(input_signal, (0, required_size - signal_length), 'constant')
    signal_length = len(padded_signal)

    if signal_length == 1:
        return padded_signal

    even_part = custom_fft(padded_signal[::2])
    odd_part = custom_fft(padded_signal[1::2])

    twiddle_factors = np.exp(-2j * np.pi * np.arange(signal_length // 2) / signal_length)

    return np.concatenate([
        even_part + twiddle_factors * odd_part,
        even_part - twiddle_factors * odd_part
    ])
```

**1.4: FFT 2**

```python
import numpy as np

def fft2(input_array):

    input_dims = np.array(input_array.shape)

    next_power_of_two = np.power(2, np.ceil(np.log2(input_dims)).astype(int))

    padded_array = np.pad(input_array,
                          ((0, next_power_of_two[0] - input_dims[0]),
                           (0, next_power_of_two[1] - input_dims[1])),
                          'constant')

    fft_result = np.zeros(padded_array.shape, dtype=np.complex128)

    for row in range(padded_array.shape[0]):
        fft_result[row, :] = custom_fft(padded_array[row, :])

    for col in range(padded_array.shape[1]):
        fft_result[:, col] = custom_fft(fft_result[:, col])

    return fft_result
```

**1.5: Comparing DFT and FFT**

```python
import numpy as np
import time
import matplotlib.pyplot as plt

array_sizes = [128, 256, 512, 1024, 2048, 4096]

dft_computation_times = []
fft_computation_times = []

for size in array_sizes:
    print(f"Testing with array size {size}...")
    signal = np.random.random(size)

    print("Computing DFT...")
    start = time.time()
    _ = custom_dft(signal)
    dft_duration = time.time() - start
    dft_computation_times.append(dft_duration)
    print("DFT completed.")

    print("Computing FFT...")
    start = time.time()
    _ = custom_fft(signal)
    fft_duration = time.time() - start
    fft_computation_times.append(fft_duration)
    print("FFT completed.")

plt.figure(figsize=(10, 6))
plt.plot(array_sizes, dft_computation_times, 'b-o', label='DFT')
plt.plot(array_sizes, fft_computation_times, 'r-o', label='FFT')
plt.xlabel('Array Size')
plt.ylabel('Computation Time (s)')
plt.title('Comparison of DFT and FFT Computation Times')
plt.grid(True)
plt.legend()
plt.yscale('log')
plt.show()

print("\nComputation Time Comparison:")
print("Size\tDFT Time (s)\tFFT Time (s)\tSpeedup Factor")
for i, size in enumerate(array_sizes):
    speedup_factor = dft_computation_times[i] / fft_computation_times[i]
    print(f"{size}\t{dft_computation_times[i]:.4f}\t{fft_computation_times[i]:.4f}\t{speedup_factor:.2f}x")
```

**1.6: 2D DFT and FFT Signals**

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_signals(signal_size=128):

    grid_x = np.linspace(-np.pi, np.pi, signal_size)
    grid_y = np.linspace(-np.pi, np.pi, signal_size)
    mesh_x, mesh_y = np.meshgrid(grid_x, grid_y)

    signal_1 = 0.5 * (1 + np.sin(mesh_x))
    signal_2 = 0.5 * (1 + np.sin(mesh_x) * np.sin(mesh_y))
    signal_3 = 0.5 * (1 + np.sin(mesh_x**2 + mesh_y**2))

    return signal_1, signal_2, signal_3

def visualize_signal_and_spectrum(input_signal, signal_title, dft_output, fft_output):
    fig, ax = plt.subplots(1, 3, figsize=(15, 5))

    signal_img = ax[0].imshow(input_signal, cmap='gray')
    ax[0].set_title(f'Original {signal_title}')
    plt.colorbar(signal_img, ax=ax[0])

    dft_mag_spectrum = np.log1p(np.abs(np.fft.fftshift(dft_output)))
    dft_img = ax[1].imshow(dft_mag_spectrum, cmap='viridis')
    ax[1].set_title('DFT Magnitude Spectrum')
    plt.colorbar(dft_img, ax=ax[1])

    fft_mag_spectrum = np.log1p(np.abs(np.fft.fftshift(fft_output)))
    fft_img = ax[2].imshow(fft_mag_spectrum, cmap='viridis')
    ax[2].set_title('FFT Magnitude Spectrum')
    plt.colorbar(fft_img, ax=ax[2])

    plt.tight_layout()
    plt.show()

signal_size = 128
signal_1, signal_2, signal_3 = generate_signals(signal_size)

signals_list = [signal_1, signal_2, signal_3]
titles_list = ['Signal = 0.5(1 + sin(x))',
               'Signal = 0.5(1 + sin(x) * sin(y))',
               'Signal = 0.5(1 + sin(x + y))']

for signal, title in zip(signals_list, titles_list):

    dft_output = dft2(signal)
    fft_output = fft2(signal)

    visualize_signal_and_spectrum(signal, title, dft_output, fft_output)
```

**1.7: 2D DFT of an image**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def analyze_color_image(image):

    blue_channel, green_channel, red_channel = cv2.split(image)

    fft_blue = fft2(blue_channel.astype(float))
    fft_green = fft2(green_channel.astype(float))
    fft_red = fft2(red_channel.astype(float))

    fft_blue_second = fft2(fft_blue)
    fft_green_second = fft2(fft_green)
    fft_red_second = fft2(fft_red)

    return (fft_blue, fft_green, fft_red), (fft_blue_second, fft_green_second, fft_red_second)

def display_color_spectra(image, first_fft, second_fft):
    fig, axes = plt.subplots(3, 3, figsize=(18, 18))

    for i, (fft_data, channel_color) in enumerate(zip(first_fft, ['Blue', 'Green', 'Red'])):
        magnitude_spectrum = np.abs(np.fft.fftshift(fft_data))
        magnitude_spectrum = np.log1p(magnitude_spectrum)
        image_display = axes[0, i].imshow(magnitude_spectrum, cmap='plasma')
        axes[0, i].set_title(f'Magnitude Spectrum (First FFT - {channel_color} Channel)')
        plt.colorbar(image_display, ax=axes[0, i])

    for i, (fft_data, channel_color) in enumerate(zip(second_fft, ['Blue', 'Green', 'Red'])):
        magnitude_spectrum = np.abs(np.fft.fftshift(fft_data))
        magnitude_spectrum = np.log1p(magnitude_spectrum)
        image_display = axes[1, i].imshow(magnitude_spectrum, cmap='plasma')
        axes[1, i].set_title(f'Magnitude Spectrum (Second FFT - {channel_color} Channel)')
        plt.colorbar(image_display, ax=axes[1, i])

    for i, (fft_data, channel_color) in enumerate(zip(second_fft, ['Blue', 'Green', 'Red'])):
        phase_spectrum = np.angle(np.fft.fftshift(fft_data))
        image_display = axes[2, i].imshow(phase_spectrum, cmap='twilight')
        axes[2, i].set_title(f'Phase Spectrum (Second FFT - {channel_color} Channel)')
        plt.colorbar(image_display, ax=axes[2, i])

    plt.tight_layout()
    plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def ifft1(X):
    N = len(X)
    if N == 1:
        return X

    X_conj = np.conjugate(X)

    even = X_conj[::2]
    odd = X_conj[1::2]

    even_ifft = ifft1(even)
    odd_ifft = ifft1(odd)

    factor = np.exp(2j * np.pi * np.arange(N) / N)
    combined = np.zeros(N, dtype=complex)
    half = N // 2

    for k in range(half):
        combined[k] = even_ifft[k] + factor[k] * odd_ifft[k]
        combined[k + half] = even_ifft[k] - factor[k] * odd_ifft[k]

    return np.real(combined) / N

reconstructed_signal = ifft1(X)

plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
plt.plot(X)
plt.title('Original Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(np.abs(X))
plt.title('Frequency Spectrum')
plt.xlabel('Frequency Bin')
plt.ylabel('Magnitude')

plt.subplot(3, 1, 3)
plt.plot(reconstructed_signal)
plt.title('Reconstructed Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
def process_fingerprint(image):
    image_float = np.float32(image)
    gaussian_blur = cv2.GaussianBlur(image_float, (5, 5), sigmaX=1.0)
    highpass_kernel = np.array([[-1, -1, -1],
                                [-1, 8, -1],
                                [-1, -1, -1]])
    image_ridges = cv2.filter2D(gaussian_blur, -1, highpass_kernel)
    contrast_enhanced = cv2.normalize(image_ridges, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
    binary_fingerprint = cv2.adaptiveThreshold(contrast_enhanced.astype(np.uint8),
                                               maxValue=255,
                                               adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                               thresholdType=cv2.THRESH_BINARY,
                                               blockSize=15,
                                               C=3)

    return binary_fingerprint

original_image = cv2.imread('fingerprint.png', cv2.IMREAD_GRAYSCALE)
processed_fingerprint = process_fingerprint(original_image)
plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.imshow(original_image, cmap='gray')
plt.title('Original Fingerprint')

plt.subplot(122)
plt.imshow(processed_fingerprint, cmap='gray')
plt.title('Processed Fingerprint')

plt.show()
```

```python
def enhance_image_with_fft_and_gaussian_smoothing(image_path):
    original_img = cv2.imread(image_path)
    rgb_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

    scale = 2
    high_res_img = cv2.resize(rgb_img, (rgb_img.shape[1] * scale, rgb_img.shape[0] * scale), interpolation=cv2.INTER_CUBIC)

    processed_channels = []

    for channel in cv2.split(high_res_img):
        frequency_domain = np.fft.fft2(channel)
        shifted_frequency = np.fft.fftshift(frequency_domain)

        rows, cols = channel.shape
        center_x, center_y = rows // 2, cols // 2
        sigma_val = 20
        x_axis = np.linspace(-cols // 2, cols // 2, cols)
        y_axis = np.linspace(-rows // 2, rows // 2, rows)
        X, Y = np.meshgrid(x_axis, y_axis)

        gaussian_filter = np.exp(-(X**2 + Y**2) / (2 * sigma_val**2))

        filtered_freq = shifted_frequency * gaussian_filter

        inv_shifted = np.fft.ifftshift(filtered_freq)
        spatial_domain_img = np.fft.ifft2(inv_shifted)
        abs_spatial_img = np.abs(spatial_domain_img)

        normalized_img = cv2.normalize(abs_spatial_img, None, 0, 255, cv2.NORM_MINMAX)

        processed_channels.append(normalized_img.astype(np.uint8))

    final_img = cv2.merge(processed_channels)

    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(rgb_img)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(final_img)
    plt.title('Enhanced Image with Gaussian Smoothing')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

image_path = '/content/google.jpeg'
enhance_image_with_fft_and_gaussian_smoothing(image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def restore_color_image(img):
    blue, green, red = cv2.split(img)
    def clean_channel(channel):
        blurred_channel = cv2.GaussianBlur(channel, (5, 5), 0)
        edge_preserved = cv2.bilateralFilter(blurred_channel, 9, 75, 75)
        contrast_enhancer = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        enhanced_channel = contrast_enhancer.apply(edge_preserved)

        return enhanced_channel

    blue_enhanced = clean_channel(blue)
    green_enhanced = clean_channel(green)
    red_enhanced = clean_channel(red)

    cleaned_image = cv2.merge([blue_enhanced, green_enhanced, red_enhanced])

    return cleaned_image

original_image = cv2.imread('cart.jpg')
restored_image = restore_color_image(original_image)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(restored_image, cv2.COLOR_BGR2RGB))
plt.title('Restored Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
gaussian_kernel = (9, 9)
morphological_element = (13,13)
min_area = 5000
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def count_coins(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, gaussian_kernel, 0)

    _, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    kernel = np.ones(morphological_element, np.uint8)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)
    closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel, iterations=2)

    contours, _ = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    valid_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > min_area]
    num_coins = len(valid_contours)

    plt.figure(figsize=(12, 8))
    images = [img, blurred, thresh, opening, closing, cv2.drawContours(img.copy(), valid_contours, -1, (0, 255, 0), 2)]
    titles = ['Original Image', 'Blurred', 'Thresholded', 'After Opening', 'After Closing', f'Final Result: {num_coins} coins']

    for i in range(6):
        plt.subplot(2, 3, i+1)
        if i == 0 or i == 5:
            plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
        else:
            plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])

    plt.tight_layout()
    plt.show()

    return num_coins
```

```python
num_coins = count_coins(coin_image_path)
print(f"Number of coins detected: {num_coins}")
```

```python
import cv2
import numpy as np

img = cv2.imread('/content/human.png', cv2.IMREAD_UNCHANGED)

alpha_channel = img[:, :, 3]
_, binary = cv2.threshold(alpha_channel, 127, 255, cv2.THRESH_BINARY)

skel = np.zeros(binary.shape, np.uint8)

element = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))

done = False
while not done:
    eroded = cv2.erode(binary, element)
    temp = cv2.dilate(eroded, element)
    temp = cv2.subtract(binary, temp)
    skel = cv2.bitwise_or(skel, temp)
    binary = eroded.copy()

    zeros = cv2.countNonZero(binary)
    if zeros == 0:
        done = True
dilated_skeleton = cv2.dilate(skel, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2)))

contrast_skeleton = cv2.normalize(dilated_skeleton, None, 0, 255, cv2.NORM_MINMAX)

plt.imshow(contrast_skeleton, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

img_path = '/content/fingerprint2.png'
img = Image.open(img_path).convert('L')
grayscale_img = np.array(img) / 255.0

smoothed_img = cv2.medianBlur((grayscale_img * 255).astype(np.uint8), 13)

_, binarized_img = cv2.threshold(smoothed_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
binarized_img = binarized_img.astype(np.uint8)

closed_structure = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
closed_result = cv2.morphologyEx(binarized_img, cv2.MORPH_CLOSE, closed_structure)

skeleton = np.zeros_like(closed_result, dtype=np.uint8)
temp_img = closed_result.copy()
while cv2.countNonZero(temp_img) > 0:
    eroded = cv2.erode(temp_img, cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3)))
    dilated = cv2.dilate(eroded, cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3)))
    temp_diff = cv2.subtract(temp_img, dilated)
    skeleton = cv2.bitwise_or(skeleton, temp_diff)
    temp_img = eroded.copy()

dilated_skel = cv2.dilate(skeleton, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (1, 1)))
enhanced_skeleton = cv2.normalize(dilated_skel, None, 0, 255, cv2.NORM_MINMAX)

fig, ax = plt.subplots(1, 4, figsize=(20, 8))
ax[0].imshow(grayscale_img, cmap='gray')
ax[0].set_title("Original Grayscale Image")
ax[0].axis('off')

ax[1].imshow(smoothed_img, cmap='gray')
ax[1].set_title("Median-Filtered Image")
ax[1].axis('off')

ax[2].imshow(closed_result, cmap='gray')
ax[2].set_title("Closed (Morphology)")
ax[2].axis('off')

ax[3].imshow(enhanced_skeleton, cmap='gray')
ax[3].set_title("Skeletonized with Enhanced Contrast")
ax[3].axis('off')

plt.show()
```
