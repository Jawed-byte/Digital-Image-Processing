Function to read an image

```python
def read_image(image_path, mode=1):
    image = cv2.imread(image_path,mode)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

Function to write image to a specified path

```python
def write_image(image, output_path):

    if len(image.shape) == 2:
      cv2.imwrite(output_path, image)
    elif len(image.shape) == 3 and image.shape[2] == 3:
      cv2.imwrite(output_path, image)
    else:
        raise ValueError("Image shape is incorrect")
```

Function to display an image

```python
def display_image(image):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.show()
```

Function to resize an image

```python
def resize_image(image, width, height):
  resized_image = np.zeros((height, width, 3), dtype=np.uint8)
  original_height, original_width, _ = image.shape
  x_scale = original_width / width
  y_scale = original_height / height

  for y in range(height):
    for x in range(width):
      original_x = int(x * x_scale)
      original_y = int(y * y_scale)
      resized_image[y, x, :] = image[original_y, original_x, :]

  return resized_image
```

Function to change the brightness of an image. If we add a constant k to each pixel brightness of the image will increase. And if we substract a constant k from each pixels brightness will decrease.

```python
def change_brightness(image, k):
  image_copy = image.copy()
  m, n = image_copy.shape[:2]
  for i in range(m):
    for j in range(n):
      pixel_value = int(image_copy[i][j])
      if pixel_value + k > 255:
        image_copy[i][j] = 255
      elif pixel_value + k < 0:
        image_copy[i][j] = 0
      else:
        image_copy[i][j] = image_copy[i][j] + k
  return image_copy
```

Increasing and decreasing the brightness of an image

```python
apple_after_increasing_brightness = change_brightness(apple,100)
apple_after_decreasing_brightness = change_brightness(apple,-100)
```

Function of contrast stretching by finding minimum value pixel and maximum value pixel

```python
def contrast_stretch(image):
  image_copy = image.copy()
  m, n = image_copy.shape[:2]
  mx = 0
  mn = 255
  for i in range(m):
    for j in range(n):
      mx = max(mx,image_copy[i][j])
      mn = min(mn,image_copy[i][j])
  for i in range(m):
    for j in range(n):
      pixel_value = int(image_copy[i][j])
      image_copy[i][j] = ((pixel_value-mn)*255)/(mx-mn)
  return image_copy
```

Function to convert a RGB image to a gray-scale image

```python
def rgbtoGray(image, alpha=0.2126, beta=0.7152, gama=0.0722):
  image_copy = image.copy()
  m, n, l = image_copy.shape[:3]
  for i in range(m):
    for j in range(n):
      red = int(image_copy[i][j][2])*alpha
      green = int(image_copy[i][j][1])*beta
      blue = int(image_copy[i][j][0])*gama
      avg = (red+green+blue)/(alpha+beta+gama)
      image_copy[i][j][0] = image_copy[i][j][1] = image_copy[i][j][2] = avg
  return image_copy
```

Function to convert gray scale image to a rgb image using pseudo mapping

```python
def pseudo_colour(image):
  image_copy = image.copy()
  m, n = image_copy.shape[:2]
  image_3_channel = []
  for i in range(m):
    row = []
    for j in range(n):
      row.append([0, 0, 0])
    image_3_channel.append(row)

  for i in range(m):
    for j in range(n):
      pixel_value = int(image_copy[i][j])
      if pixel_value < 70:
        image_3_channel[i][j] = [pixel_value, 0, 0]
      elif pixel_value < 170:
        image_3_channel[i][j] = [0, pixel_value, 0]
      elif pixel_value < 230:
        image_3_channel[i][j] = [0, pixel_value, pixel_value]
      else:
        image_3_channel[i][j] = [0, 0, pixel_value]

  return np.array(image_3_channel)
```

Function to convert green screen background to a specified background in an input image

```python
def replace_green_screen(input_img, background_img):
  input_img_copy = input_img.copy()
  background_img_copy = background_img.copy()
  m, n = input_img_copy.shape[:2]
  for i in range(m):
    for j in range(n):
      if input_img_copy[i][j][1] > 200 and int(input_img_copy[i][j][1])-int(input_img_copy[i][j][0]) > 50 and int(input_img_copy[i][j][1])-int(input_img_copy[i][j][2]) > 50:
        input_img_copy[i][j] = background_img_copy[i][j]
  return input_img_copy
```

Function to read video

```python
def read_video(video_path):
    cap = cv2.VideoCapture(video_path)
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)
    cap.release()
    return frames
```

```python
video_frames = read_video('../media/input_video.mp4')
```

Function to write back the video

```python
def write_video(frames, output_path, fps=30):
  height, width, _ = frames[0].shape
  fourcc = cv2.VideoWriter_fourcc(*'mp4v')
  out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
  for frame in frames:
    out.write(frame)
  out.release()
```

```python
write_video(video_frames, '../media/output_video.mp4')
```

Function to create transition video

```python
def create_transition(img1, img2, duration_seconds, fps):
  # Resize images to a common size
  target_width = min(img1.shape[1], img2.shape[1])
  target_height = min(img1.shape[0], img2.shape[0])
  img1_resized = resize_image(img1, target_width, target_height)
  img2_resized = resize_image(img2, target_width, target_height)

  num_frames = duration_seconds * fps
  frames = []

  for i in range(num_frames):
    alpha = i / (num_frames - 1)
    blended_frame = (img1_resized * (1 - alpha) + img2_resized * alpha).astype(np.uint8)
    frames.append(blended_frame)

  return frames
```

Creating transition video from transition image1 to transition image2
